# Data Preprocessing

```python
import pandas as pd
```

## Import Data

```python
urlemdat = 'https://github.com/Themaoyc/MDA/blob/main/Data/emdat%20heatwave.csv?raw=true'
urlgdp='https://github.com/Themaoyc/MDA/blob/main/Data/GDP.csv?raw=true'
urlpopulation='https://github.com/Themaoyc/MDA/blob/main/Data/population.csv?raw=true'
df = pd.read_csv(urlemdat)
GDP = pd.read_csv(urlgdp)
population= pd.read_csv(urlpopulation)
```

```python
##choose the useful column from emdat
df=df[["Dis No","Year","Country","Start Year","Start Month","Start Day","End Year",
      "End Month","End Day","Dis Mag Value","Total Deaths","CPI"]]
df
```

## Merge data 

```python
GDP
```

```python
GDP.drop(GDP.columns[2:GDP.columns.get_loc('1993')], axis=1,inplace=True)
GDP
```

```python
GDP.groupby(['Scale']).count() 
```

### The scale is the same(million dollars) for all the records, so we can drop this colum for simplicity

```python
GDP.drop(['Scale'], axis=1,inplace=True)
```

```python
GDP=pd.melt(GDP, id_vars=['Country'], value_vars=GDP.columns[1:30], 
            var_name='Year',value_name='GDP(million dollars)', col_level=None)
GDP
```

### The contry in different tables may be recorded in different ways(i.e. Czech Rep. & Czech Republic), we should rewrite them before merge.

```python
df.loc[df['Country'] == 'United States of America (the)','Country']= 'United States'
df.loc[df['Country'] == 'Netherlands (the)','Country']= 'Netherlands, The'
df.loc[df['Country'] == 'Macedonia (the former Yugoslav Republic of)','Country']= 'North Macedonia, Republic of'
df.loc[df['Country'] == 'Czech Republic (the)','Country']= 'Czech Rep.'
df.loc[df['Country'] == 'Korea (the Republic of)','Country']= 'Korea, Rep. of'
df.loc[df['Country'] == 'Sudan (the)','Country']= 'Sudan'
df.loc[df['Country'] == "Korea (the Democratic People's Republic of)",'Country']= 'Dem. Rep. Korea '
df.loc[df['Country'] == 'Bolivia (Plurinational State of)','Country']= 'Bolivia'
df.loc[df['Country'] == 'Russian Federation (the)','Country']= 'Russian Federation'
df.loc[df['Country'] == 'United Kingdom of Great Britain and Northern Ireland (the)','Country']= 'United Kingdom'
df.loc[df['Country'] == 'Egypt','Country']= 'Egypt, Arab Rep. of'
df.loc[df['Country'] == 'Slovakia','Country']= 'Slovak Rep.'
df.loc[df['Country'] == 'Slovenia','Country']= 'Slovenia, Rep. of'
df.loc[df['Country'] == 'Serbia','Country']= 'Serbia, Rep. of'
df.loc[df['Country'] == 'Serbia','Country']= 'Serbia, Rep. of'
```

```python
GDP['Year']=GDP['Year'].astype(int)
merge1 = pd.merge(df, GDP, on=['Year','Country'],how='left')
merge1
```

### Merge with the population data

```python
population
```

```python
##similar operations as gdp
population.drop(population.columns[1:population.columns.get_loc('1993')], axis=1,inplace=True)
population=population.rename(columns={'Country Name':'Country'})
population=pd.melt(population, id_vars=['Country'], value_vars=population.columns[1:30], 
            var_name='Year',value_name='Population', col_level=None)
population['Year']=population['Year'].astype(int)
population.loc[population['Country'] == 'Netherlands','Country']= 'Netherlands, The'
population.loc[population['Country'] == 'Turkiye','Country']= 'Turkey'
population.loc[population['Country'] == 'Serbia','Country']= 'Serbia, Rep. of'
population.loc[population['Country'] == 'Slovak Republic','Country']= 'Slovak Rep.'
population.loc[population['Country'] == 'Slovenia','Country']= 'Slovenia, Rep. of'
population.loc[population['Country'] == 'North Macedonia','Country']= 'North Macedonia, Republic of'
population.loc[population['Country'] == 'Egypt, Arab Rep.','Country']= 'Egypt, Arab Rep. of'
population.loc[population['Country'] == 'Czech Republic','Country']= 'Czech Rep.'
population.loc[population['Country'] == 'Korea, Rep.','Country']= 'Korea, Rep. of'
population['Year']=population['Year'].astype(int)
population.loc[population['Country'] == 'Egypt, Arab Rep.','Country']= 'Egypt, Arab Rep. of'
merge2 = pd.merge(merge1, population, on=['Year','Country'],how='left')
merge2
```

## Dealing with the missing data

```python
df.isna().sum()
```

### For the missing Start Day and End Day, the value 1 is given.

```python
df=df.fillna({"Start Day":1,"End Day":1})
df
```

### The missing Dis Mag Value(the max temperature during the heatwave) is replaced by the average of other heatwaves.

```python
Tavg=df['Dis Mag Value'].mean()
df=df.fillna({"Dis Mag Value":Tavg})
df[["Start Day","End Day","Dis Mag Value"]]=df[["Start Day","End Day","Dis Mag Value"]].astype(int)
df
```

### For the missing value of total deaths, we drop these records.

```python
population=population[["Population"]]=population[["Population"]].astype(int)
```
