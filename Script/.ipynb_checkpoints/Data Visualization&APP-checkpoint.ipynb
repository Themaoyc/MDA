```python
df_map =pd.DataFrame(df) 
countryname= df_map['Country']
df_map['Countryshort'] = df_map['Country']
df_map.loc[df_map['Country'] == 'United States of America (the)','Countryshort']= 'United States'
df_map.loc[df_map['Country'] == 'Netherlands (the)','Countryshort']= 'Netherlands'
df_map.loc[df_map['Country'] == 'Macedonia (the former Yugoslav Republic of)','Countryshort']= 'Macedonia'
df_map.loc[df_map['Country'] == 'Czech Republic (the)','Countryshort']= 'Czech Rep.'
df_map.loc[df_map['Country'] == 'Korea (the Republic of)','Countryshort']= 'Korea'
df_map.loc[df_map['Country'] == 'Sudan (the)','Countryshort']= 'Sudan'
df_map.loc[df_map['Country'] == "Korea (the Democratic People's Republic of)",'Countryshort']= 'Dem. Rep. Korea '
df_map.loc[df_map['Country'] == 'Bolivia (Plurinational State of)','Countryshort']= 'Bolivia'
df_map.loc[df_map['Country'] == 'Russian Federation (the)','Countryshort']= 'Russia'
df_map.loc[df_map['Country'] == 'United Kingdom of Great Britain and Northern Ireland (the)','Countryshort']= 'United Kingdom'
df_map.loc[df_map['Country'] == 'Bosnia and Herzegovina','Countryshort']= 'Bosnia and Herz.'

#calculate the times of heatwave in a country in the same year


df_map['cy']=df_map['Countryshort']+df_map['Year'].apply(str)
times=df_map["cy"].value_counts().to_dict()

df_map['times']=1
df_map.loc[df_map['cy'] == 'Japan2015','times']= 3
df_map.loc[df_map['cy'] == 'Belgium2019 ','times']= 3
df_map.loc[df_map['cy'] == 'United Kingdom2019','times']= 2
df_map.loc[df_map['cy'] == 'Germany2019 ','times']= 2
df_map.loc[df_map['cy'] == 'United States2006 ','times']= 2
df_map.loc[df_map['cy'] == 'Australia1994 ','times']= 2
df_map.loc[df_map['cy'] == 'Japan2019 ','times']= 2
df_map.loc[df_map['cy'] == 'India2019 ','times']= 2
df_map.loc[df_map['cy'] == 'France2019 ','times']= 2
df_map.head()
```

```python
hwmap = Timeline()
for i in range(1993, 2022):
        df_all_layer = df_map.loc[df_map['Year']==i]
        map0 = (
        Map()
        .add("", [list(z) for z in zip(df_all_layer['Countryshort'], df_all_layer['times'])], "world",is_map_symbol_show=False)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Heatwave Map of Year{}".format(i)),
            visualmap_opts=opts.VisualMapOpts(max_=3,min_=0),
        )
    )
        hwmap.add(map0, "{}Year".format(i))
hwmap.render("map.html")
```

```python
import pandas as pd
import dash
from bokeh.embed import file_html
from dash import html
from dash import dcc
import plotly.graph_objects as go
import plotly.express as px
from dash.dependencies import Input, Output
from skimage import io

from scipy.signal import savgol_filter


import networkx as nx
from community import community_louvain
from toolz.dicttoolz import valmap
from bokeh.palettes import Blues8
from bokeh.io import output_notebook
from bokeh.plotting import figure
from bokeh.models import Range1d, Circle, ColumnDataSource,MultiLine, NodesAndLinkedEdges, LabelSet
from bokeh.plotting import from_networkx
from bokeh.resources import CDN

data = pd.read_csv("data/belgium_temp.csv", parse_dates=True)
mort = pd.read_csv("data/belgium_deaths.csv", parse_dates=True)
dailytemp = pd.read_csv("data/belgium_daily_temp.csv", parse_dates=True)
full = pd.read_csv("data/belgium_full.csv", parse_dates=True)
forecast = pd.read_csv("data/belgium_forecast.csv", parse_dates=True)
belgium_2022 = pd.read_csv("data/belgium_2022.csv", parse_dates=True)
forecast_for_prediction = pd.read_csv("data/forecast_for_prediction.csv", parse_dates=True)
nodes = pd.read_csv("data/nodes_df.csv", parse_dates=True)
topic_mod = pd.read_csv("data/topic_modelling.csv", parse_dates=True)
df_graph_net = pd.read_csv("data/crosstab_5_sample.csv", parse_dates=True)
df_emotion = pd.read_csv("data/month_emo_count.csv", parse_dates=True)
pd.options.mode.chained_assignment = None

from preprocessing.mortality.year_select import year_select
from scripts.Poisson import Poisson

app = dash.Dash(__name__,suppress_callback_exceptions=True)

server = app.server

app.layout = html.Div([

    dcc.Tabs(
        id="tabs",
        value='tab-1',
        parent_className='custom-tabs',
        className='custom-tabs-container',
        children=[
            dcc.Tab(
                label='Visualization',
                value='tab-1',
                className='custom-tab',
                selected_className='custom-tab--selected'
            ),
            dcc.Tab(
                label='Death Predictions',
                value='tab-2',
                className='custom-tab',
                selected_className='custom-tab--selected'
            ),
            dcc.Tab(
                label='Emotion Classification',
                value='tab-3',
                className='custom-tab',
                selected_className='custom-tab--selected'
            ),
            dcc.Tab(
                label='Topic Modelling',
                value='tab-4',
                className='custom-tab',
                selected_className='custom-tab--selected'
            ),
            dcc.Tab(
                label='WordCloud',
                value='tab-5',
                className='custom-tab',
                selected_className='custom-tab--selected'
            ),
            dcc.Tab(
                label='Hashtags Co-occurrence Network - Centrality Measures',
                value='tab-6',
                className='custom-tab',
                selected_className='custom-tab--selected'
            ),
            dcc.Tab(
                label='Hashtags Co-occurrence Network and Communities',
                value='tab-7',
                className='custom-tab',
                selected_className='custom-tab--selected'
            ),

        ]),
    html.Div(id='tab-content'),
])

@app.callback(Output('tab-content', 'children'),
              Input('tabs', 'value'))

def render_content(tab):
    if tab == 'tab-1':

        return html.Div([

            html.H3('Temperature and Mortality'),

            dcc.Graph(id="graph"),
            dcc.Slider(id='year-slider',


                    step=None,
                    marks={
                        2017: '2017',
                        2018: '2018',
                        2019: '2019',
                        2020: '2020',
                        2021: '2021',
                        2022: '2022'
                    },
                    value=2020
            ),
            dcc.RadioItems(id="avg-onoff",
                options=["Daily temperatures", "Monthly average temperatures"],
                value='Daily temperatures',

            ),
        ])

    elif tab == 'tab-2':
        return html.Div([
            html.H3('SARIMA and Poisson Regression'),

            dcc.Graph(id="death-pred", style={'textAlign': 'center'}),

            dcc.RadioItems(id="death-toggle",
                           options=["Summer 2022","Yearly"],
                           value='Summer 2022',
                           ),
        ])

    elif tab == 'tab-3':
        return html.Div([
            html.H3('Emotion Classification'),

            dcc.RadioItems(id="emotion-toggle",
                         value='Emotion',
                         options=['Emotion'] ),


            dcc.Graph(id="emotion", style={'textAlign': 'center'})

        ])

    elif tab == 'tab-4':
        return html.Div([
            html.H3('Topic Modelling'),
            html.H4('Heatwave'),
            html.Label('Choose a Topic'),
            html.Br(),  # Break

            dcc.Dropdown(id="lda-toggle",
                           options=[
                             #  {'label': i, 'value': i}
                              # for i in topic_mod['level_0'].unique()
                              {'label': "Recreational Activities", 'value': 'Topic 1'},  # label: front-end; # value: needs to align with the excel values
                              {'label': "Tips", 'value': 'Topic 2'},
                              {'label': "High Temperature", 'value': 'Topic 3'},
                           ],
                           value='Topic 1',
                           ),
            dcc.Graph(id="lda", style={'textAlign': 'center'})

        ])

    elif tab == 'tab-5':
        return html.Div([
            html.H3('WordCloud'),
            html.H4('Top 30 words of the topics'),
            html.Label('Choose a Topic'),
            html.Br(),  # Break

            dcc.Dropdown(id="wordcloud-toggle",
                           options=[
                             #  {'label': i, 'value': i}
                              # for i in topic_mod['level_0'].unique()
                              {'label': "Recreational Activities", 'value': 'Topic 1'},  # label: front-end; # value: needs to align with the excel values
                              {'label': "Tips", 'value': 'Topic 2'},
                              {'label': "High Temperature", 'value': 'Topic 3'},
                           ],
                           value='Topic 1',
                           ),
            dcc.Graph(id="wordcl", style={'textAlign': 'center'})

        ])

    elif tab == 'tab-6':
        return html.Div([
            html.H3('Top 10 Hashtags'),
            html.H4('Centrality Algorithms'),
            html.Label('Choose a Measure'),

            dcc.Dropdown(id="nodes-toggle",
                         options=[
                             {'label': "Centrality Measure", 'value': 'degree'},
                             {'label': "PageRank", 'value': 'pageranks'},
                             {'label': "Betweenness", 'value': 'betweenness'},
                         ],
                         value='degree',
                         multi=False,
                         clearable=False
                         ),

            dcc.Graph(id="nodes", style={'textAlign': 'center'})

        ])

    elif tab == 'tab-7':
        return html.Div([
            html.H3('Hashtags Co-occurence Network'),
            html.H4('Sample of 100 Hashtags'),

            dcc.RadioItems(id="network-toggle",
                         value='Hashtags',
                         options=['Hashtags'] ),


            html.Iframe(
                id='network',
                srcDoc=None,  # here is where we will put the graph we make
                style={'border-width': '5', 'width': '100%', 'height': '500px'})

        ])


@app.callback(
    Output("graph", "figure"),
    [Input("year-slider", "value"), Input("avg-onoff", "value")])


def display_(year_value, radio_value):

    df = year_select(year_value, data, mort, dailytemp)


    temp = df['AverageTemperature']
    temp = savgol_filter(temp, 60, 2)

    temp2 = belgium_2022['temp']
    temp2 = savgol_filter(temp2, 155, 2)

    fig = go.Figure()

    if radio_value=="Daily temperatures":

        if year_value == 2022:
            fig.add_trace(
                go.Scatter(x=belgium_2022['date'], y=belgium_2022['temp'], name="Daily Temp.", line=dict(color="#1887ab"))
            )
        else:
            fig.add_trace(
                go.Scatter(x=df['Date'], y=df['DailyTemp'], name="Daily Temp.", line=dict(color="#1887ab"))

            )

            fig.add_trace(
                go.Scatter(x=df['Date'], y=df['Deaths'], name="Deaths", line=dict(color="#FBC543"),
                           yaxis="y2")
            )

    elif radio_value=="Monthly average temperatures":
        if year_value == 2022:
            fig.add_trace(
                go.Scatter(x=belgium_2022['date'], y=temp2, name="Monthly Temp.", line=dict(color="#1887ab"))
            )
        else:
            fig.add_trace(
                go.Scatter(x=df['Date'], y=temp, name="Monthly Temp.", line=dict(color="#1887ab"))
            )
            fig.add_trace(
                go.Scatter(x=df['Date'], y=df['Deaths'], name="Deaths", line=dict(color="#FBC543"),
                           yaxis="y2")
            )


    fig.update_layout(

        yaxis=dict(
            title="<b>Temperature (°C)</b>",
            titlefont=dict(
                color="#1f77b4"

            ),
            tickfont=dict(
                color="#1887ab"
            )
        ),
        yaxis2=dict(
            title="<b>Deaths</b>",
            titlefont=dict(
                color="#1f77b4"
            ),
            tickfont=dict(
                color="#FBC543"
            ),
            anchor="free",
            overlaying="y",
            side="right",
            position=1.0
        )
    )

    fig.update_layout(
        title_text="Belgium",
        width=1900,
        height = 600

    )

    #fig.update_xaxes(title_text="<b>Date</b>", color="#1f77b4")
    fig.update_xaxes(title_text="<b>Date</b>", color="#1f77b4")

    fig.update_layout(autotypenumbers='convert types')

    fig.layout.template = "plotly_white+ygridoff"
    return fig

@app.callback(
    Output("death-pred", "figure"),
    Input("death-toggle", "value"))

def display2(death_toggle_value):
    model = Poisson(full)
    model.fit()
    pred = model.predict(forecast_for_prediction)
    predsum = sum(pred)

    dates = belgium_2022[(belgium_2022['date'] >= '2022-06-01') & (belgium_2022['date'] <= '2022-09-30')]
    dates = dates['date']

    if death_toggle_value == "Summer 2022":

        belgium_2022_temp = belgium_2022[(belgium_2022['date'] >= '2022-06-01') & (belgium_2022['date'] <= '2022-09-30')]
        belgium_2022_temp = belgium_2022_temp['temp']

        fig = go.Figure()

    # fig.add_trace(
    #    go.Bar(x=df['Month'], y=df['Deaths'], name="Deaths by month", marker_color='#1887ab')
    # )

        fig.add_trace(
            go.Scatter(x=dates, y=belgium_2022_temp, name="Daily Temp.", line=dict(color="#1887ab"))
        )

        fig.add_trace(
            go.Scatter(x=dates, y=pred, name="Deaths", line=dict(color="#FBC543"), yaxis="y2")
        )

        fig.update_layout(
            yaxis=dict(
                title="<b>Deaths</b>",
                titlefont=dict(
                    color="#1f77b4"

                ),
                tickfont=dict(
                    color="#1f77b4"
                )
            )

        )

        fig.update_xaxes(title_text="<b>Month</b>", color="#1f77b4")

        fig.layout.template = "plotly_white+ygridoff"

        fig.update_layout(
            title_text="Belgium",
            height=600

        )

        fig.update_layout(
            yaxis=dict(
                title="<b>Temperature (°C)</b>",
                titlefont=dict(
                    color="#1887ab"

                ),
                tickfont=dict(
                    color="#1887ab"
                )
            ),
            yaxis2=dict(
                title="<b>Deaths</b>",
                titlefont=dict(
                    color="#1f77b4"
                ),
                tickfont=dict(
                    color="#FBC543"
                ),
                anchor="free",
                overlaying="y",
                side="right",
                position=1.0
            )
        )

        return fig

    elif death_toggle_value == "Yearly":

        df = [["2017",sum(full[(full.Year==2017) & ((full.Month==6) | (full.Month==7) | (full.Month==8) | (full.Month==9))]['Deaths'])],
              ["2018",sum(full[(full.Year==2018) & ((full.Month==6) | (full.Month==7) | (full.Month==8) | (full.Month==9))]['Deaths'])],
              ["2019",sum(full[(full.Year==2019) & ((full.Month==6) | (full.Month==7) | (full.Month==8) | (full.Month==9))]['Deaths'])],
              ["2020",sum(full[(full.Year==2020) & ((full.Month==6) | (full.Month==7) | (full.Month==8) | (full.Month==9))]['Deaths'])],
              ["2021",sum(full[(full.Year==2021) & ((full.Month==6) | (full.Month==7) | (full.Month==8) | (full.Month==9))]['Deaths'])],
              ["2022 (Predicted)",predsum]]
        df = pd.DataFrame(df, columns=['Year', 'Deaths'])

        fig = go.Figure()

        fig.add_trace(
            go.Bar(x=df['Year'], y=df['Deaths'], name="Deaths by year", marker_color='#1887ab')
        )

        fig.update_layout(
            yaxis=dict(
                title="<b>Deaths</b>",
                titlefont=dict(
                    color="#1f77b4"

                ),
                tickfont=dict(
                    color="#1f77b4"
                )
            )

        )

        fig.layout.template = "plotly_white+ygridoff"

        fig.update_xaxes(title_text="<b>Year</b>", color="#1f77b4")
        #color="#F0F8FF"
        return fig



@app.callback(
    Output("emotion", "figure"),
    Input("emotion-toggle", "value"))

def display3(lda_toggle_value):

    fig_emotion = px.line(df_emotion, x="created_at", y="id", color='Emotion',
                       labels={'created_at': 'Date', 'id':'Count'})

    return fig_emotion


@app.callback(
    Output("lda", "figure"),
    Input("lda-toggle", "value"))

def display4(lda_toggle_value):
    filtered_topics_mod = topic_mod[topic_mod['level_0'] == lda_toggle_value]
    topic_fig = px.bar(filtered_topics_mod,
                       x = 'tag',
                       y = 'weight',
                       color = 'tag',
                       labels={'tag': 'Words', 'weight':'Weights'})
    topic_fig.update_layout(showlegend=False)
    return topic_fig

@app.callback(
    Output("nodes", "figure"),
    Input("nodes-toggle", "value"))

def display5(nodes_toggle_value):
    
    centrality_top10 = nodes[['hashtag', 'degree']].sort_values(by='degree', ascending=False)[1:11]

    if nodes_toggle_value == "degree":
        nodes_fig = px.bar(centrality_top10,
                           x='hashtag',
                           y='degree',
                           color='hashtag',
                           labels={'hashtag': 'Hashtag', 'degree': 'Centrality'})
        nodes_fig.update_layout(showlegend=False)

        return nodes_fig

    elif nodes_toggle_value == "pageranks":
        pagerank_top10 = nodes[['hashtag', 'pageranks']].sort_values(by='pageranks', ascending=False)[1:11]

        nodes_fig = px.bar(pagerank_top10,
                           x='hashtag',
                           y='pageranks',
                           color='hashtag',
                           labels={'hashtag': 'Hashtag', 'pageranks': 'PageRank'})
        nodes_fig.update_layout(showlegend=False)


        return nodes_fig

    elif nodes_toggle_value == "betweenness":

        betweenness_top10 = nodes[['hashtag', 'betweenness']].sort_values(by='betweenness', ascending=False)[1:11]

        nodes_fig = px.bar(betweenness_top10,
                           x='hashtag',
                           y='betweenness',
                           color='hashtag',
                           labels={'hashtag': 'Hashtag', 'betweenness': 'Betweenness'})
        nodes_fig.update_layout(showlegend=False)


        return nodes_fig

@app.callback(
    Output("network", "srcDoc"),
    Input("network-toggle", "value"))

def display6(nodes_toggle_value):
    df_graph_net_1 = df_graph_net.copy()
    g = nx.from_pandas_edgelist(df_graph_net_1, source='source', target='target', edge_attr='weight')
    louvain = community_louvain.best_partition(g)
    louvain_sorted = dict(sorted(louvain.items(), key=lambda item: item[1]))
    new_color = (
        '#FDCD90',
        '#29788E',
        '#FBFCBF',
        '#482374',
        '#404387',
        '#345E8D',
        '#29788E',
        '#208F8C',
        '#22A784',
        '#42BE71',
        '#79D151',
        '#BADE27',
        '#FDE724')

    def new_func(x):
        if x == 0:
            return '#FDCD90'
        elif x == 1:
            return '#FBFCBF'
        elif x == 2:
            return '#482374'
        elif x == 3:
            return '#404387'
        elif x == 4:
            return '#345E8D'
        elif x == 5:
            return '#29788E'
        elif x == 6:
            return '#208F8C'
        elif x == 7:
            return '#22A784'
        elif x == 8:
            return '#42BE71'
        elif x == 9:
            return '#79D151'
        elif x == 10:
            return '#BADE27'
        elif x == 11:
            return '#FDE724'
        elif x == 12:
            return '#29788E'
        elif x == 13:
            return '#42BE71'

    modularity_class = louvain_sorted
    modularity_color = valmap(new_func, louvain)

    degrees = dict(nx.degree(g))
    nx.set_node_attributes(g, name='degree', values=degrees)

    # Add modularity class and color as attributes from the network above
    nx.set_node_attributes(g, modularity_class, 'modularity_class')
    nx.set_node_attributes(g, modularity_color, 'modularity_color')

    number_to_adjust_by = 1
    adjusted_node_size = dict([(node, degree + number_to_adjust_by) for node, degree in nx.degree(g)])
    nx.set_node_attributes(g, name='adjusted_node_size', values=adjusted_node_size)

    # Choose colors for node and edge highlighting
    node_highlight_color = 'white'
    edge_highlight_color = 'black'

    # Choose attributes from G network to size and color by — setting manual size (e.g. 10) or color (e.g. 'skyblue') also allowed
    size_by_this_attribute = 'adjusted_node_size'
    color_by_this_attribute = 'modularity_color'

    # Pick a color palette — Blues8, Reds8, Purples8, Oranges8, Viridis8
    color_palette = Blues8

    # Choose a title!
    #title = 'Hashtags Co-occurence Network'

    # Establish which categories will appear when hovering over each node
    HOVER_TOOLTIPS = [
        ("Character", "@index"),
        ("Degree", "@degree"),
        ("Modularity Class", "@modularity_class"),
        ("Modularity Color", "$color[swatch]:modularity_color"),
    ]

    # Create a plot — set dimensions, toolbar, and title
    plot = figure(tooltips=HOVER_TOOLTIPS,
                  tools="pan,wheel_zoom,save,reset", active_scroll='wheel_zoom',
                  x_range=Range1d(-10.1, 10.1), y_range=Range1d(-10.1, 10.1), #title=title,
                  height=900,width=1490
                  )

    # Create a network graph object
    # https://nx.github.io/documentation/nx-1.9/reference/generated/nx.drawing.layout.spring_layout.html
    network_graph = from_networkx(g, nx.spring_layout, scale=10, center=(0, 0))

    # Set node sizes and colors according to node degree (color as category from attribute)
    network_graph.node_renderer.glyph = Circle(size=size_by_this_attribute, fill_color=color_by_this_attribute)
    # Set node highlight colors
    network_graph.node_renderer.hover_glyph = Circle(size=size_by_this_attribute, fill_color=node_highlight_color,
                                                     line_width=2)
    network_graph.node_renderer.selection_glyph = Circle(size=size_by_this_attribute, fill_color=node_highlight_color,
                                                         line_width=2)

    # Set edge opacity and width
    network_graph.edge_renderer.glyph = MultiLine(line_alpha=0.3, line_width=1)
    # Set edge highlight colors
    network_graph.edge_renderer.selection_glyph = MultiLine(line_color=edge_highlight_color, line_width=2)
    network_graph.edge_renderer.hover_glyph = MultiLine(line_color=edge_highlight_color, line_width=2)

    # Highlight nodes and edges
    network_graph.selection_policy = NodesAndLinkedEdges()
    network_graph.inspection_policy = NodesAndLinkedEdges()

    plot.renderers.append(network_graph)

    # Add Labels
    x, y = zip(*network_graph.layout_provider.graph_layout.values())
    node_labels = list(g.nodes())
    source = ColumnDataSource({'x': x, 'y': y, 'name': [node_labels[i] for i in range(len(x))]})
    labels = LabelSet(x='x', y='y', text='name', source=source, background_fill_color='white', text_font_size='10px',
                      background_fill_alpha=.7)
    plot.renderers.append(labels )
    html_bokeh = file_html(plot, CDN)


    return html_bokeh

@app.callback(
    Output("wordcl", "figure"),
    Input("wordcloud-toggle", "value"))


def display7(wordcloud_toggle_value):

    if wordcloud_toggle_value == "Topic 1":

        img = io.imread(
            'data/top_0.jpg')
        fig = px.imshow(img).update_xaxes(showticklabels = False).update_yaxes(showticklabels = False)

        fig.update_layout(
            width=1400,
            height=600
        )
        return fig

    elif wordcloud_toggle_value == "Topic 2":

        img = io.imread(
            'data/top_1.jpg')
        fig = px.imshow(img).update_xaxes(showticklabels = False).update_yaxes(showticklabels = False)

        fig.update_layout(
            width=1400,
            height=600
        )

        return fig

    elif wordcloud_toggle_value == "Topic 3":

        img = io.imread(
            'data/top_2.jpg')
        fig = px.imshow(img).update_xaxes(showticklabels = False).update_yaxes(showticklabels = False)
        fig.update_layout(
            width=1400,
            height=600
        )
        return fig


if __name__ == '__main__':
    app.run_server(port=9999,debug=True)
```
